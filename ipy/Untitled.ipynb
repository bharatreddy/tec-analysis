```python
def convert_to_datetime(row):
        """ 
        Given a datestr and a time string for 
        losVel DF convert to a python datetime obj.
        """
        import datetime
        datecolName="date"
        timeColName="time"
        currDateStr = str( int( row[datecolName] ) )
    #     return currDateStr
        if row[timeColName] < 10:
            currTimeStr = "000" + str( int( row[timeColName] ) )
        elif row[timeColName] < 100:
            currTimeStr = "00" + str( int( row[timeColName] ) )
        elif row[timeColName] < 1000:
            currTimeStr = "0" + str( int( row[timeColName] ) )
        else:
            currTimeStr = str( int( row[timeColName] ) )
        return datetime.datetime.strptime( currDateStr\
                        + ":" + currTimeStr, "%Y%m%d:%H%M" )

geoCoordsRbspFile = "/home/bharat/Documents/code/rbsp-geo-pos/rbspPosSatA.txt"
geoRbspMagnsphrDF = pandas.read_csv(geoCoordsRbspFile, \
                             delim_whitespace=True, header=None)

geoRbspMagnsphrDF.columns = [ "date", "time", "alt", "lat", "lon" ]

geoRbspMagnsphrDF = geoRbspMagnsphrDF[ geoRbspMagnsphrDF["date"] == 20121230 ].reset_index(drop=True)

geoRbspMagnsphrDF["datetime"] = geoRbspMagnsphrDF.apply( \
                                    convert_to_datetime, axis=1 )
# geoRbspMagnsphrDF["date"] = [ x.strftime("%Y%m%d") for x in geoRbspMagnsphrDF["datetime"] ]
# geoRbspMagnsphrDF["time"] = [ x.strftime("%H%M") for x in geoRbspMagnsphrDF["datetime"] ]

geoRbspMagnsphrDF["year"] = [ x.year for x in geoRbspMagnsphrDF["datetime"] ]
geoRbspMagnsphrDF["month"] = [ x.month for x in geoRbspMagnsphrDF["datetime"] ]
geoRbspMagnsphrDF["day"] = [ x.day for x in geoRbspMagnsphrDF["datetime"] ]
geoRbspMagnsphrDF["hour"] = [ x.hour for x in geoRbspMagnsphrDF["datetime"] ]
geoRbspMagnsphrDF["min"] = [ x.minute for x in geoRbspMagnsphrDF["datetime"] ]
outCols = [ "year", "month", "day", "hour", "min", "alt", "lat", "lon" ]
geoRbspMagnsphrDF.to_csv( "../data/muhammad-magnLoc-geoCoords.csv",index=None, sep=" ", columns=outCols )
geoRbspMagnsphrDF.head()
```
